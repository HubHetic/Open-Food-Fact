# Cleaning the data

```python
#drop na des colonnes qui ns intéresse en prio
#image_url : l'url permettant d'acceder à notre image,
#product_name : le nom de notre produit
#brands : la marque associé au produit
```

```python
#We need image url and product name to be full, else we won't able to do anything
df.dropna(inplace=True, subset=['image_url','product_name'])
```

```python
#After browsing the data, we noticed that we people add informations about
#brand and quantity, the pictures are well taken
df.dropna(inplace=True, subset=['brands', 'quantity'])
```

```python
#Killiweb is one of the funder of Yuka, they did a pre cleaning of the data
df = df[df["creator"] == "kiliweb"]
```

```python
#In the case you want everything in french
df = df[df["countries_tags"]== "en:france"]
```

```python
#columns who are 100% empty
df.drop('states',1, inplace=True)
df.drop('states_tags',1, inplace=True)
df.drop('states_en',1, inplace=True)
df.drop('pnns_groups_2',1, inplace=True)
df.drop('pnns_groups_1',1, inplace=True)
```

```python
#create a dictionnary to know the % of filling of each colulmns
t = dict(zip(df.columns.values.tolist(), ((np.array(df.isna().sum().tolist()) / df.shape[0]) ))) 
for i in t :
    print(f"{i} : {round(t[i],5)}")
```

```python
#suppriles les colonnes avec plus de 90% de NaN
a = [i[0] for i in t if i[1] > 0.95]
df.drop(a ,1, inplace=True)
    
```

# Split to train and test

```python
#Here we want to get all the product name with more than 3 occurence
#Have one dataset test with one of those occurences, the train with the rest
#create a copy of our first df
df1 = df.copy()
df1['count'] = 1
```

```python
#we groupby name,  sort it, and convert it to a dict
a = df1.groupby(['product_name'])['count'].agg('sum')
a.sort_values( ascending=False)
dic_dup = a.to_dict()
```

```python
#we create a new colum repet where we give him the number of repetition of our dictionnarie 
df1['repet'] = df1['product_name'].map(lambda x : dic_dup[x])
df1.sort_values(['repet'])    
```

```python
#here we create a new dataframe where the value of the column repet is >=3
#wich means we ll get a df where you can find the same product name at least 3 times
df2 = df1[df1['repet'] >= 3]
```

```python
df2.shape
```

```python
#we drop duplicates to have only one occurence of the productname, it s our test data
df_test = df2.drop_duplicates(subset=['product_name'])
df_test.shape
```

```python
#we exclude the data from the df_test from df_2, it will gives us the data train
df_train = df1.copy()
df_train = df_train.merge(df_test, indicator='i', how='outer').query('i == "left_only"').drop('i', 1)
df_train.shape
```
